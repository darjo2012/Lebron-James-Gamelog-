**NOTE**: There are several more issues with this dataset:
1. The column 'W/L' needs to be trimmed to show either an 'L' or a 'W'. 
2. There are games where Lebron did not play, I think the best approach is to just remove the games in which he didn't play. Write a program to remove rows from the dataset that has 'Inactive' or 'Did not dress'.
3. FT% and FG% have extra null values for the days Lebron didn't make a single free throw or 3 pt. Do we keep the values null or replace it with 0?

```python
#The column 'W/L' needs to be trimmed to show either an 'L' or a 'W'.
lebron['W/L'] = lebron['W/L'].astype(str).str[:1]
```

```python
#There are games where Lebron did not play, I think the best approach is to just remove the games in which he didn't play. Write a program to remove rows from the dataset that has 'Inactive' or 'Did not dress'.
lebron.isnull().sum()
```

```python
lebron.drop(lebron[lebron['GS'] == 'Inactive'].index, inplace = True)
```

```python
lebron.drop(lebron[lebron['GS'] == 'Did Not Play'].index, inplace = True)
```

```python
lebron.drop(lebron[lebron['GS'] == 'Not With Team'].index, inplace = True)
```

```python
lebron.drop(lebron[lebron['GS'] == 'Did Not Dress'].index, inplace = True)
```

```python
#FT% and FG% have extra null values for the days Lebron didn't make a single free throw or 3 pt. Do we keep the values null or replace it with 0?
lebron['3P%'].fillna(0, inplace = True)
```
